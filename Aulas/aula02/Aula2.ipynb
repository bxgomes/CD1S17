```python
casA = planCas[tvplano == "A"] # Pessoas casadas que possuem o Plano A
casB = planCas[tvplano == "B"] # Pessoas casadas que possuem o Plano B

lenCasA = len(casA)
lenCasB = len(casB)

print("Número de pessoas casadas que possuem o plano A:", lenCasA)
print("Número de pessoas casadas que possuem o plano B:", lenCasB)
```

```python
tvmtsA = casA[tvsatis == "Muito Satisfeito"]
tvsatA = casA[tvsatis == "Satisfeito"]
tvindA = casA[tvsatis == "Indiferente"]
tvinsA = casA[tvsatis == "Insatisfeito"]
tvmtiA = casA[tvsatis == "Muito Insatisfeito"]

lenMtsA = len(tvmtsA)
lenSatA = len(tvmtsA)
lenTSA = lenMtsA + lenSatA

lenIndA = len(tvindA)

lenMtsA = len(tvmtiA)
lenInsA = len(tvinsA)
lenTIA = lenMtsA + lenInsA

print("Pessoas casados que estão satisfeitas com o plano A:", lenTSA)
print("Pessoas casados que estão indiferentes com o plano A:", lenIndA)
print("Pessoas casados que estão insatisfeitas com o plano A:", lenTIA)
```

```python
tvmtsB = casB[tvsatis == "Muito Satisfeito"]
tvsatB = casB[tvsatis == "Satisfeito"]
tvindB = casB[tvsatis == "Indiferente"]
tvinsB = casB[tvsatis == "Insatisfeito"]
tvmtiB = casB[tvsatis == "Muito Insatisfeito"]

lenMtsB = len(tvmtsB)
lenSatB = len(tvmtsB)
lenTSB = lenMtsB + lenSatB

lenIndB = len(tvindB)

lenMtsB = len(tvmtiB)
lenInsB = len(tvinsB)
lenTIB = lenMtsB + lenInsB

print("Pessoas casados que estão satisfeitas com o plano B:", lenTSB)
print("Pessoas casados que estão indiferentes com o plano B:", lenIndB)
print("Pessoas casados que estão insatisfeitas com o plano B:", lenTIB)
```

```python
PorSatA = lenTSA/lenCasA
PorIndA = lenIndA/lenCasA
PorInsA = lenTIA/lenCasA

print("Porcentagem dos casados que estão satisfeitos com o plano A:", PorSatA*100, "%")
print("Porcentagem dos casados que estão indiferentes com o plano A:", PorIndA*100, "%")
print("Porcentagem dos casados que estão insatisfeitos com o plano A:", PorInsA*100, "%")
```

```python
PorSatB = lenTSB/lenCasB
PorIndB = lenIndB/lenCasB
PorInsB = lenTIB/lenCasB

print("Porcentagem dos casados que estão satisfeitos com o plano B:", PorSatB*100, "%")
print("Porcentagem dos casados que estão indiferentes com o plano B:", PorIndB*100, "%")
print("Porcentagem dos casados que estão insatisfeitos com o plano B:", PorInsB*100, "%")
```
