### 4)

```python
mu1 = dados.A.mean()
mu2 = dados.B.mean()

sigma1 = (dados.A.var())**0.5
sigma2 = (dados.B.var())**0.5
```

```python
fig = plt.figure(figsize = (13, 10))

plt.subplot(221)
plot = dados.A.plot.hist(bins = faixa, title = 'Renda no Bairro A', figsize = (13, 10), alpha = 0.5, color = 'b', normed = 1)
plt.plot(df.A, stats.norm.pdf(df.A, loc = mu1, scale = sigma1), lw = 1.5, alpha = 0.6, color = 'r')
plt.ylabel('Frequência absoluta')
plt.xlabel('Renda (em mil reais)')

plt.subplot(222)
plot = dados.B.plot.hist(bins = faixa, title = 'Renda no Bairro B', figsize = (13, 10), alpha = 0.5, color = 'g', normed = 1)
plt.plot(df.B, stats.norm.pdf(df.B, loc = mu2, scale = sigma2), lw = 1.5, alpha = 0.6, color = 'r')
plt.ylabel('Frequência absoluta')
plt.xlabel('Renda (em mil reais)')
plt.ylim(0, 0.7)
```

```python
fra = []
for i in range (1, 501):
    fra.append ((i - 0.5)/500)
```

```python
qA = [] # quantis teoricos da distribuição normal A
for i in fra:
    qA.append(stats.norm.ppf(i, loc = mu1, scale = sigma1))

qB = [] # quantis teoricos da distribuição normal B
for i in fra:
    qB.append(stats.norm.ppf(i, loc = mu2, scale = sigma2))
```

```python
plt.figure(figsize = (12, 6))

plt.subplot(121)
plt.plot(df.A, qA, 'o', color = 'r')
plt.plot((13, 18), (13, 18), color = 'b')
plt.title('Bairro A')
plt.xlabel('Renda A')
plt.ylabel('Quantis Teóricos da Distribuição Normal')

plt.subplot(122)
plt.plot(df.B, qB, 'o', color = 'r')
plt.plot((8, 20), (8, 20), color = 'b')
plt.title('Bairro B')
plt.xlabel('Renda B')
plt.ylabel('Quantis Teóricos da Distribuição Normal')
```

```python
paA = [] # Probabilidade Acumulada A
for i in df.A:
    paA.append(stats.norm.cdf(i, loc = mu1, scale = sigma1))
    
paB = [] # # Probabilidade Acumulada A
for i in df.B:
    paB.append(stats.norm.cdf(i, loc = mu2, scale = sigma2))
```

```python
plt.figure(figsize = (12, 6))

plt.subplot(121)
plt.plot(df.A, fra, lw = 1.5, color = 'g', alpha = 0.6)
plt.plot(df.A, paA, lw = 1.5, color = 'b', alpha = 0.6)
plt.ylabel('Frequência relativa aculmulada')
plt.xlabel('Renda A')
plt.title('Bairro A')

plt.subplot(122)
plt.plot(df.B, fra, lw = 1.5, color = 'g', alpha = 0.6)
plt.plot(df.B, paB, lw = 1.5, color = 'b', alpha = 0.6)
plt.ylabel('Frequência relativa acumulada')
plt.xlabel('Renda B')
plt.title('Bairro B')
```

Ao analisar os gráficos obtidos pecebemos que as rendas da região A podem ser representadas por uma distribuição normal melhor do que a região B, que possuirá uma margem de erro maior.

### 5 A)

```python
print('Probabilidade de uma família do bairro A ter uma renda maior que R$ 17 mil:', 
      1 - (stats.norm.cdf(17, loc = mu1, scale = sigma1)), '\n')

print('Probabilidade de uma família do bairro B ter uma renda maior que R$ 17 mil:', 
      1 - (stats.norm.cdf(17, loc = mu2, scale = sigma2)))
```

### 5 B)

```python
# P(x<16|x>15) = P(15<x<16)/P(x>15) 

A1 = stats.norm.cdf(16, loc = mu1, scale = sigma1)
A2 = stats.norm.cdf(15, loc = mu1, scale = sigma1)
A3 = stats.norm.cdf(19, loc = mu1, scale = sigma1)

B1 = stats.norm.cdf(16, loc = mu2, scale = sigma2)
B2 = stats.norm.cdf(15, loc = mu2, scale = sigma2)
B3 = stats.norm.cdf(19, loc = mu2, scale = sigma2)


print('Probabilidade de uma família do bairro A ter uma renda inferior a R$ 16 mil, dado que ela seja maior do que R$ 15 mil:', 
      ((A1 - A2)/(A3 - A2)), '\n')

print('Probabilidade de uma família do bairro B ter uma renda inferior a R$ 16 mil, dado que ela seja maior do que R$ 15 mil:', 
      ((B1 - B2)/(B3 - B2)))
```

### 5 C)

```python
A4 = stats.norm.ppf(0.95, loc = mu1, scale = sigma1)
B4 = stats.norm.ppf(0.95, loc = mu2, scale = sigma2)

print('A menor renda familiar dos 5% das famílias que mais ganham no bairro A (em mil reias) é:', A4, '\n')
print('A menor renda familiar dos 5% das famílias que mais ganham no bairro B (em mil reias) é:', B4)
```
